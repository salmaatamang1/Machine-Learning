```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pd.read_csv("titanic dataset.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
#cleaning of dataset
sum(df.duplicated(subset='PassengerId'))==0
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True,axis=0, how='any')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
auto=df[['PassengerId','Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked']]
auto
```

```python
gender=pd.get_dummies(auto['Sex'],drop_first=True)
auto=pd.concat([auto,gender],axis=1)
auto
```

```python
auto.drop(['Sex'],axis=1,inplace=True)
```

```python
auto
```

```python
auto.info()
```

```python
cabin=pd.get_dummies(auto['Cabin'],drop_first=True)
auto=pd.concat([auto,cabin],axis=1)
auto
```

```python
embark=pd.get_dummies(auto['Embarked'],drop_first=True)
auto=pd.concat([auto,embark],axis=1)
auto
```

```python
auto.drop(['Cabin','Embarked'],inplace=True,axis=1)
auto
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
train,test=train_test_split(auto,test_size=0.3,random_state=100)
train
```

```python
test.columns
```

```python
train.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
arr=['PassengerId','Survived','Pclass','Age','SibSp','Parch', 'Fare']
train[arr]=scaler.fit_transform(train[arr])
```

```python
train
```

```python
plt.figure(figsize=(150,150))
sn.heatmap(train.corr(),annot=True)
plt.show()
```

```python
y_train=train.pop('Survived')
x_train=train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
lm=LinearRegression()
lm.fit(x_train,y_train)
rfe=RFE(lm,n_features_to_select=10)
rfe=rfe.fit(x_train,y_train)
x_train.columns
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
col=x_train.columns[rfe.support_]
col
```

```python
x_train_rfe=x_train[col]
```

```python
plt.figure(figsize=(150,100))
sn.heatmap(x_train_rfe.corr(),annot=True)
plt.show()
```

```python
#adding constant
import statsmodels.api as sm
x_train_rfe=sm.add_constant(x_train_rfe)
x_train_rfe
```
